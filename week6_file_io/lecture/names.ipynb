```python
name = input("What is your name? ")

with open("names.txt", "a") as file:
    file.write(f"{name}\n")
```

```python
with open("names.txt", "r") as file:
    names = file.read().splitlines()

print("Names in file:")
for name in names:
    print(f"Hi {name}")
```

```python
names = []

with open("names.txt") as file:
    for line in file:
        names.append(line.strip())

for name in sorted(names):
    print(f"Hi {name}!")
```
