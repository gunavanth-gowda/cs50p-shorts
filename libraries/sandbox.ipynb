```python
import random

number_of_tosses = int(input("How many times would you like to toss the coin? "))

cards = ["Ace", "2", "3", "4", "5", "6", "7", "8", "9", "10", "Jack", "Queen", "King"]
random.shuffle(cards)

for i in range(random.randint(1, number_of_tosses)):
    print(f"Toss {i + 1}: {random.choice(['Heads', 'Tails'])}")

for card in cards:
    print(f"Card: {card}")
```

```python
import requests
import json

song_name = input("What is the song name? ")

response = requests.get("https://itunes.apple.com/search?entity=song&limit=1&term="+ song_name).json()

print(json.dumps(response, indent=2))
```

```python
import requests

def main():
    artist = input("Which artist would you like to search for? ")

    try:
        content = requests.get(
            "https://api.artic.edu/api/v1/artworks/search",
            {"q": artist}
        )
        content.raise_for_status()
    except requests.HTTPError:
        print("An error occurred while fetching the data.")
        return

    for artwork in content.json()["data"]:
        print(f"* {artwork["title"]}")

main()
```
