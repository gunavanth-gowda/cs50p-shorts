```python
import re

code = input("Hexadecimal color code: ")

pattern = r"#[abcdefABCDEF0123456789]"
match = re.search(pattern, code)

if match:
    print(f"Valid. Matched with {match.group()}")
else:
    print(f"Invalid. Not matched")
```

```python
import re

code = input("Hexadecimal color code: ")

pattern = r"#[abcdefABCDEF0123456789]{6}"
match = re.search(pattern, code)

if match:
    print(f"Valid. Matched with {match.group()}")
else:
    print(f"Invalid. Not matched")
```

```python
import re

code = input("Hexadecimal color code: ")

pattern = r"^#[abcdefABCDEF0123456789]{6}^"
match = re.search(pattern, code)

if match:
    print(f"Valid. Matched with {match.group()}")
else:
    print(f"Invalid. Not matched")
```

```python
import re

code = input("Hexadecimal color code: ")

pattern = r"^#[a-fA-F0-9]{6}^"
match = re.search(pattern, code)

if match:
    print(f"Valid. Matched with {match.group()}")
else:
    print(f"Invalid. Not matched")
```
