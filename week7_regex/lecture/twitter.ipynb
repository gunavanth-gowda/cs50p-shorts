```python
url = input("enter your URL: ").strip()

user_name = url.replace("https://twitter.com/", "")
print(f"Username: {user_name}")
```

```python
url = input("enter your URL: ").strip()

user_name = url.removeprefix("https://twitter.com/")
print(f"Username: {user_name}")
```

```python
import re

url = input("enter your URL: ").strip()

user_name = re.sub(r".*(https?://)?(www\.)?twitter\.com/", "", url)
print(f"Username: {user_name}")
```

```python
import re

url = input("enter your URL: ").strip()

if matches := re.search(
    r"^(?:https?://)?(?:www\.)?twitter\.com/(.+)$", url, re.IGNORECASE
):
    print(f"Username: {matches.group(1)}")
```
