```python
name = input("enter your name: ").strip()

if "," in name:
    last_name, first_name = name.split(",")
    name = f"{first_name.strip()} {last_name}"

print(f"hello, {name}")
```

```python
import re

name = input("enter your name: ").strip()

matches = re.search(r"^(.+), (.+)$", name)

if matches:
    name = f"{matches.group(2)} {matches.group(1)}"

print(f"hello, {name}")
```

```python
import re

name = input("enter your name: ").strip()

matches = re.search(r"^(.+), ?(.+)$", name)

if matches:
    name = f"{matches.group(2)} {matches.group(1)}"

print(f"hello, {name}")
```

```python
import re

name = input("enter your name: ").strip()

matches = re.search(r"^(.+), *(.+)$", name)

if matches:
    name = f"{matches.group(2)} {matches.group(1)}"

print(f"hello, {name}")
```
